```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
years_x = [1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
total_y = [1243, 1543, 1619, 1831, 1960, 2310, 2415, 2270, 1918]
```

```python
plt.plot(years_x, total_y);
```

```python
plt.plot(years_x, total_y)
plt.ylabel('MtCO2/yr')
plt.title('CO2 emissions from electricity production - US')
plt.show()
```

```python
coal_y = [823, 1136, 1367, 1547, 1660, 1927, 1983, 1827, 1352]
gas_y = [171, 200, 166, 175, 228, 280, 319, 399, 529]
```

```python
#start a figure
plt.figure(figsize=(10,3))

#First subplot
plt.subplot(1,2,1)
plt.title("coal vs gas")
plt.xlim(1975,2015)
plt.ylim(0,2400)
plt.plot(years_x, coal_y)
plt.plot(years_x, gas_y)

#Second subplot
plt.subplot(1,2,2)
plt.xlim(1975,2015)
plt.title("all")
plt.plot(years_x, total_y)

plt.suptitle("US Co2 emissions")
plt.show()
```

```python
#destructuring

fig, axs = plt.subplots(2,2, figsize=(10,6))

axs[0,0].plot(years_x, coal_y, label="coal")
axs[0,0].plot(years_x, gas_y, label = "gas")
axs[0,0].set_title('coal vs. gas')
axs[0,0].legend()

axs[0,1].plot(years_x, total_y, c='black')
axs[0,1].set_title('all energies')

axs[1,0].plot(years_x, total_y, c='yellow')
axs[1,0].set_title('all energies')

axs[1,1].plot(years_x, coal_y, label="coal")
axs[1,1].plot(years_x, gas_y, label = "gas")
axs[1,1].set_title('coal vs. gas')
axs[1,1].legend()

plt.show()
```

```python
# plot method with pandas dataframe

df1 = pd.DataFrame({'coal':coal_y, 'gas':gas_y}, index=years_x)
df2 = pd.DataFrame({'total':total_y}, index=years_x)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
df1.plot(ax=ax1)
df2.plot(ax=ax2)
```
